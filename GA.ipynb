```python
#N-Queens Problem 

import random

mutation = 0.75 #mutation rate
popSize = 50 #individuals in each generation
#N=8


def Gridisize(chromo):  #Chess Board Printing
  grid = [[' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' '],
          [' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' '],[' ',' ',' ',' ',' ',' ',' ',' ']] 
  ind = 0
  for gene in chromo: 
      grid[gene][ind]= 'Q'
      ind+=1
  return grid


def FitnessFunc(chrome):  #Each chromosome is evaluated through this fitness function
    non_attacking_queens = 0   
    for row in range(8):
        col = chrome[row]    
        for iter in range(8):  
            if iter != row and chrome[iter] != col: #check for itself and horizontal
                if iter + chrome[iter] != row + col and  iter - chrome[iter] != row - col: #diagonal checks
                    non_attacking_queens += 1
    return non_attacking_queens//2


def mate(p1,p2):    #one point crossover
  r = random.randint(0,7)
  mated = p1[0:r] + p2[r:]
  return mated

def swap(chrome1,chrome2):
  temp = chrome1
  x = chrome2
  chrome2 = temp


def mutate(chrome):   #two random indexes values are swapped and changing one of the index to a randomize value
  swap(chrome[random.randint(0,7)],chrome[random.randint(0,7)])  #swapping random values in chromosome
  chrome[random.randint(0,7)]=random.randint(0,7)                #randomly changing any index
  return chrome


def GeneticAlgo(population):  #Parent Function
  new_pop = []
  population.sort(reverse = True)
  new_size=popSize//4
  new_pop = population[0:new_size]  #new population

  for i in range(len(new_pop)):       #random parent selection
      parent1 = population[random.randint(0,new_size//4)][1]
      parent2 = population[random.randint(0,new_size//4)][1]
      child = mate(parent1,parent2)
      if random.random() < mutation:
          child = mutate(child)
      new_pop.append((FitnessFunc(child),child))
      if(new_pop[-1][0] == 0):  #if fitness becomes zero
          break
  return new_pop
      

#chromosome builder function
def chromoArr():
    row=[0,1,2,3,4,5,6,7]
    return [random.choice(row),random.choice(row),random.choice(row),random.choice(row),random.choice(row),random.choice(row),random.choice(row),random.choice(row)]

#Code Start:
chromosomes=[]
for i in range (popSize):
  chromosomes.append(chromoArr());
PopFit=[]
for i in range(popSize):
  PopFit.append((FitnessFunc(chromosomes[i]),chromosomes[i]))

generation = 0

#Algorithm with 1000 iterations
for x in range(1000):
  PopFit = GeneticAlgo(PopFit)
  print("Generation no {}: ".format(x))
  for p in PopFit:
      print(p)
  
PopFit.sort(reverse=True)

#Printing the best solution
print("Best Solution:: {}".format(PopFit[0]))
result=Gridisize(PopFit[0][1])
for r in result:
  print(r)

```
